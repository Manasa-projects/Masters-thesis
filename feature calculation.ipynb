```python
#finding permutations for string AGCT for a hexamer
from itertools import product
strg1='AGCT'
allcomb=list(product(list(strg1),repeat=6))
```

```python
#allcomb elements are a tuple,so convert to string
def convertTuple(tup):
    strg=''
    for item in tup:
        strg=strg+item
    return strg

allcombstr=''
allcomblist=[]
for i in range(0,len(allcomb)):
    allcombstr=convertTuple(allcomb[i])
    allcomblist.append(allcombstr)
    
```

```python
#function to count the frequency of hexamers
def freqcount(seqq,window,step):
    allcombfreq=[]
    for i in range(0,len(allcomblist)):
        allcombfreq.append([allcomblist[i],0])
    
    for i in range(0,len(seqq)-window,step):
        u1=seqq[i:i+6]
        flagi=0
        for j in range(0,len(allcomblist)):
            if u1==allcomblist[j]:
                flagi=flagi+1
       
        allcombfreq[allcomblist.index(u1)][1]=flagi
        
        
            
    return allcombfreq        
```

```python


#numeric sequence for each pct

final_numberspct=[]
for i in range(0,len(pctstrings)):
    numbers=[]
    for j in lncrnastrings[i]:
        if j=='A':
            numbers.append(0.1260)
            
        if j=='G':
            numbers.append(0.0806)
            
        if j=='C':
            numbers.append(0.1340)
            
        if j=='T':
            numbers.append(0.1335)
            
    final_numberspct.append(numbers)
    
    

    
    
import scipy
import numpy
import math

fft_valuespct=[]
for i in range(0,len(final_numberspct)):
    f=scipy.fft(final_numberspct[i])
    
    fft_valuespct.append(f)


abs_finalpct=[]
for i in range(0,len(fft_valuespct)):
    abs_values=[]
    for j in range(0,len(fft_valuespct[i])):
        f1=(abs(j))^2
        abs_values.append(f1)
        
    abs_finalpct.append(abs_values)
    
    
average_powerpct=[]
for i in range(0,len(abs_finalpct)):
    
    sums=0
    for j in range(0,len(abs_finalpct[i])):
        sums=sums+j
    
    #sum=sum/len(abs_final[i])
    
    average_powerpct.append(sums)
    
    
#signal at n/3 position

signal_n3pct=[]
for i in range(0,len(abs_finalpct)):
    length=len(abs_finalpct[i])
    j=int(length/3)
    value=abs_finalpct[i][j]
    signal_n3pct.append(value)


    
    
#snr

snrpct=[]
for i in range(0,len(signal_n3pct)):
    div=signal_n3pct[i]/average_powerpct[i]
    snrpct.append(div)
    



#quantile statistics


#sorted_list=abs_final
for i in range(0,len(abs_finalpct)):
    
    abs_finalpct[i].sort(reverse=True)

    
#calculating top 10% for each sequence

top10perpct=[]
for i in range(0,len(abs_finalpct)):
    l=len(abs_finalpct[i])
    p=int(0.1*l)
    top10perpct.append(abs_finalpct[i][0:p])
    

    
import numpy as np

max_valuepct=[]
min_valuepct=[]
Q1pct=[]
Q2pct=[]

for i in range(0,len(top10perpct)):
    ma=max(top10perpct[i])
    max_valuepct.append(ma)
    mi=min(top10perpct[i])
    min_valuepct.append(mi)
    q1=np.quantile(top10perpct[i],0.25)
    Q1pct.append(q1)
    q2=np.quantile(top10perpct[i],0.5)
    Q2pct.append(q2)
    
```

# Sequence based features for pct-ORF length and coverage

```python
#orf length and coverage

#find substrings in pctstrings

import textwrap
import time

sub=3
start_codon=['ATG']
stop_codon=['TAA','TAG','TGA']


#dividing whole string into 3 codons
x4pct=[]
for i in range(0,len(pctstrings)):
    x=textwrap.wrap(pctstrings[i],3)
    x3=[]
    for j in range(0,len(x)):
        if x[j]==start_codon[0]:
            for k in range(j,len(x)):
                if x[k]==stop_codon[0] or x[k]==stop_codon[1] or x[k]==stop_codon[2]:
                    x1=x[j:k+1]
                    x2=''.join(x1)
                    break
            x3.append(x2)
            j=k+1
    x4pct.append(x3)
    
    
#dividing whole string into 3 codons except first nucleotide

pct1=[]
for i in range(0,len(pctstrings)):
    a=pctstrings[i][1:len(pctstrings[i])]
    pct1.append(a)
    
x5pct=[]
for i in range(0,len(pct1)):
    x=textwrap.wrap(pct1[i],3)
    x3=[]
    for j in range(0,len(x)):
        if x[j]==start_codon[0]:
            for k in range(j,len(x)):
                if x[k]==stop_codon[0] or x[k]==stop_codon[1] or x[k]==stop_codon[2]:
                    x1=x[j:k+1]
                    x2=''.join(x1)
                    break
            x3.append(x2)
            j=k+1
    x5pct.append(x3)
    
    
#dividing whole string into 3 codons except first two nucleotides

pct2=[]
for i in range(0,len(pctstrings)):
    a=pctstrings[i][2:len(pctstrings[i])]
    pct2.append(a)
    
x6pct=[]
for i in range(0,len(pct2)):
    x=textwrap.wrap(pct2[i],3)
    x3=[]
    for j in range(0,len(x)):
        if x[j]==start_codon[0]:
            for k in range(j,len(x)):
                if x[k]==stop_codon[0] or x[k]==stop_codon[1] or x[k]==stop_codon[2]:
                    x1=x[j:k+1]
                    x2=''.join(x1)
                    break
            x3.append(x2)
            j=k+1
    x6pct.append(x3)
    

#calculate length os all orfs for each sequence and get the max length and coverage
h4pct=[]
for i in range(0,len(x4pct)):
    h1=[]
    for j in range(0,len(x4pct[i])):
        h=len(x4pct[i][j])
        h1.append(h)
    h4pct.append(h1)
    
h5pct=[]
for i in range(0,len(x5pct)):
    h1=[]
    for j in range(0,len(x5pct[i])):
        h=len(x5pct[i][j])
        h1.append(h)
    h5pct.append(h1)
    
h6pct=[]
for i in range(0,len(x6pct)):
    h1=[]
    for j in range(0,len(x6pct[i])):
        h=len(x6pct[i][j])
        h1.append(h)
    h6pct.append(h1)
    
m4pct=[]
for i in range(0,len(h4pct)):
    m1=max(h4pct[i],default=0)#to deal with empty lists
    m4pct.append(m1)

m5pct=[]
for i in range(0,len(h5pct)):
    m1=max(h5pct[i],default=0)
    m5pct.append(m1)

m6pct=[]
for i in range(0,len(h6pct)):
    m1=max(h6pct[i],default=0)
    m6pct.append(m1)
    
#orf length calculation

max_lengthpct=[]
for i in range(0,len(m4pct)):
    mdash=max(m4pct[i],m5pct[i],m6pct[i])
    max_lengthpct.append(mdash)



#orf coverage calculation

coveragepct=[]
for i in range(0,len(pctstrings)):
    g=len(pctstrings[i])
    c=max_lengthpct[i]/g
    coveragepct.append(c)
    
```

# Structure based features-MFE,UP frequency

```python
#get rna strings from the given dna strings

rnastringspct=pctstrings
dnastringspct=pctstrings


temp1pct=[]
for i in range(0,len(rnastringspct)):
    temp=list(rnastringspct[i])
    for j in range(0,len(temp)):
        
        
        if temp[j]=='A':
            temp[j]='a'
            
        if temp[j]=='G':
            temp[j]='g'
            
        if temp[j]=='C':
            temp[j]='c'
            
        if temp[j]=='T':
            temp[j]='u'
        
    t1=''.join(temp)
    temp1pct.append(t1)
        

temp2pct=[]
for i in range(0,len(dnastringspct)):
    tempd=list(dnastringspct[i])
    for j in range(0,len(tempd)):
        
        
        if tempd[j]=='A':
            tempd[j]='a'
            
        if tempd[j]=='G':
            tempd[j]='g'
            
        if tempd[j]=='C':
            tempd[j]='c'
            
        if tempd[j]=='T':
            tempd[j]='t'
        
    t2=''.join(tempd)
    temp2pct.append(t2)
    
rnapct=temp1pct
dnapct=temp2pct

#run rna_fold from viennarna package to get secondary structure

import subprocess
from subprocess import Popen, PIPE 
 

sspct=[]
for i in range(0,len(rnapct)):
    p = subprocess.Popen('RNAfold.exe', stdin=PIPE, stdout=PIPE)
    ans = p.communicate(pctstrings[i].encode())
    sspct.append(ans)
    
#convert bytes to string type
ss1pct=[]
for i in range(0,len(sspct)):
    t1=sspct[i][0].decode('utf-8')
    ss1pct.append(t1)
    
loc1pct=[]
for i in range(0,len(ss1pct)):
    result=ss1pct[i].find(' ')
    loc1pct.append(result)
    
db1pct=[]
for i in range(0,len(ss1pct)):
    r2=ss1pct[i][loc1pct[i]+2:len(ss1pct[i])-3]
    db1pct.append(float(r2))
    


#get the dot bracket notation

locpct=[]
for i in range(0,len(ss1pct)):
    x1=ss1pct[i].find('.')
    x2=ss1pct[i].find('(')
    x3=ss1pct[i].find(')')
    lst=[x1,x2,x3]
    y=min(lst)
    x4=ss1pct[i].find(' ')
    locpct.append([y,x4])
    
dbpct=[]
for i in range(0,len(ss1pct)):
    r1=ss1pct[i][locpct[i][0]:locpct[i][1]]
    dbpct.append(r1)
    

    
#paired-unpaired sequence

puseqpct=dbpct

u2pct=[]
for i in range(0,len(puseqpct)):
    u1=list(puseqpct[i])
    for j in range(0,len(puseqpct[i])):
        if u1[j]=='.':
            u1[j]='U'
        if u1[j]==')' or u1[j]=='(':
            u1[j]='P'
    u3=''.join(u1)
    u2pct.append(u3)
    
    
# paired-unpaired 'UP' frequency
flag_UPpct=[]
for i in range(0,len(u2pct)):
    flagu=0
    for j in range(0,len(u2pct[i])-1):
        #g=j+1
        if u2pct[i][j]=='U' and u2pct[i][j+1]=='P': #and g!=len(u2[i])-1:
            flagu=flagu+1
    flag_UPpct.append(flagu/len(u2pct[i]))
            
            
```

```python
#acguD and acgu-ACGU sequences of pct sequences

#getting acgu-ACGU sequence
newseqpct=rnapct
newseq1pct=[]
for i in range(0,len(rnapct)):
    new=list(newseqpct[i])
    for j in range(0,len(rnapct[i])):
        if rnapct[i][j]=='a' and u2pct[i][j]=='P':
            new[j]='A'
        
        if rnapct[i][j]=='g' and u2pct[i][j]=='P':
            new[j]='G'
            
        if rnapct[i][j]=='c' and u2pct[i][j]=='P':
            new[j]='C'
            
        if rnapct[i][j]=='u' and u2pct[i][j]=='P':
            new[j]='U'
            
        if u2pct[i][j]=='U':
            new[j]=rnapct[i][j]
            
    w=''.join(new)
    newseq1pct.append(w)
            
        
#getting acguD sequence
newseqpct=rnapct
newseq2pct=[]
for i in range(0,len(rnapct)):
    new=list(newseqpct[i])
    for j in range(0,len(rnapct[i])):
        if u2pct[i][j]=='U':
            new[j]='D'
            
        if u2pct[i][j]=='P':
            new[j]=rnapct[i][j]
            
    w=''.join(new)
    newseq2pct.append(w)
```
